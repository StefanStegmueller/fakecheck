```python
import spacy
from spacy import displacy
from nltk import Tree
from nltk.draw import TreeView
import copy
```

```python
nlp = spacy.load("DATA/en_core_web_sm-2.2.5/en_core_web_sm/en_core_web_sm-2.2.5")
doc1 = nlp("I ate the nice fish with a spoon")
displacy.render(doc1, style='dep', jupyter=True)
```

```python
doc2 = copy.deepcopy(doc1)
doc2[5].head = doc2[4]
displacy.render(doc2, style='dep', jupyter=True)
```

```python

svg1 = displacy.render(doc1, style="dep", jupyter=False)
svg2 = displacy.render(doc2, style="dep", jupyter=False)

f = open("doc1.svg", "w", encoding="utf-8")
f.write(svg1)
f.close()

f = open("doc2.svg", "w", encoding="utf-8")
f.write(svg2)
f.close()
```
